1. Write a program for the following variant of the singly linked list. Write
all the functionalities.
Traverse, insert at beginning, insert at last, insert at some position,
delete at beginning, delete at last, delete at some position.
The list should look like as follows:
– Each row take values from the range 100i + 1 to 100(i + 1) for

1 ≤ i ≤ 10.

– The row become empty if no value in that range is present.

– The range values are ordered top to bottom.

Insert and delete is based on the renge values. Suppose one wants to
insert 156 at the beginning. Then it will be inserted at the beginning
of second row.

If one wants to insert at beginning 666. Since the row is not present, a
new row is created after third row and the number is inserted.
If any number outsider of the range is given then it will print an warning
message that the number is not a valid number.

```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None
        self.down = None
```

```python
class TwoDim_List:
    def __init__(self):
        self.head = None
        # self.length = 0
        # self.depth = 0
    
    def insertAtLast(self, data):
        curr = self.head
        node = Node(data)

        if curr != None:
            while curr.down and curr.down.data // 100 < data // 100:
                curr = curr.down

            if curr.data // 100 != data // 100:
                if curr.down == None:
                    curr.down = node
                else:
                    if curr.down.data // 100 > data // 100:
                        node.down = curr.down
                        curr.down = node
                    else:
                        curr = curr.down
                        while curr.next != None:
                            curr = curr.next
                        curr.next = node
            else:
                while curr.next != None:
                    curr = curr.next
                curr.next = node            
        else:
            self.head = node

    def insertAtFirst(self, data):
        curr = self.head
        prevFirst = None
        node = Node(data)

        if curr != None:
            while curr.down and curr.down.data // 100 < data // 100:
                prevFirst = curr
                curr = curr.down
                
            if curr.data // 100 != data // 100:
                if curr.down == None:
                    curr.down = node
                else:
                    if curr.down.data // 100 > data // 100:
                            node.down = curr.down
                            curr.down = node
                    else:
                        downNode = curr.down
                        node.next = downNode
                        node.down = downNode.down
                        curr.down = node
                        downNode.down = None
            else:
                if prevFirst:
                    prevFirst.down = node
                else:
                    self.head = node
                node.next = curr
                node.down = curr.down
                curr.down = None
                    
        else:
            self.head = node

    def insertAtNth(self, data, n):
        node = Node(data)
        if n == 1:
            self.insertAtFirst(node)
        elif n == self.length + 1:
            self.insertAtLast(node)
        elif n <= self.length:
            curr = self.head
            i = 1
            while i < n - 1:
                curr = curr.next
                i += 1
                
            node.next = curr.next
            curr.next = node

    def deleteFirst(self):
        self.head = self.head.next
    
    def deleteLast(self):
        curr = self.head
        prev = None
        while curr.next != None:
            prev = curr
            curr = curr.next
        
        prev.next = None
        

    def deleteAtN(self, n):
        if n == 1:
            self.deleteFirst()
        elif n == self.length:
            self.deleteLast()
        elif n < self.length:
            i = 1
            prev = None
            curr = self.head
            while i < n:
                prev = curr
                curr = curr.next
                i += 1
            prev.next = curr.next
            

    def traverse(self):
        curr = self.head
        first = self.head
        while first != None:
            while curr != None:
                print(curr.data, end = " -> ")
                curr = curr.next
            
            first = first.down
            curr = first
            print("\n|")
```
